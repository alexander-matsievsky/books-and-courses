## 10.1 Removing the Legend

> Sometimes a legend is redundant, or it is supplied in another graph that will be displayed with the current one. In these cases, it can be useful to remove the legend from a graph.

> When a variable is mapped to `fill`, the default scale used is `scale_fill_discrete()` (equivalent to `scale_fill_hue()`), which maps the factor levels to colors that are equally spaced around the color wheel.

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  guides(fill = FALSE)

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  scale_fill_discrete(guide = FALSE)

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  theme(legend.position = "none")
```

## 10.2 Changing the Position of a Legend

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  theme(legend.position = "top")

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  theme(
    legend.background = element_blank(),
    legend.justification = c(1.25, -.25),
    legend.position = c(1, 0),
  )
```

## 10.3 Changing the Order of Items in a Legend

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  scale_fill_brewer(limits = c("trt1", "trt2", "ctrl"), palette = "Pastel1")
```

## 10.4 Reversing the Order of Items in a Legend

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  guides(fill = guide_legend(reverse = TRUE))
```

## 10.5 Changing a Legend Title

> It’s also possible to set the title of the legend in the scale specification. Since legends and axes are both guides, this works the same way as setting the title of the x- or y-axis.

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  labs(fill = "Condition")

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  guides(fill = guide_legend("Condition"))
```

```r
qplot(ageYear, heightIn, color = sex, size = weightLb, data = gcookbook::heightweight) +
  scale_size_continuous(range = c(1, 4)) +
  labs(color = "Male/Female", size = "Weight\n(in pounds)")

qplot(ageYear, heightIn, color = sex, shape = sex, data = gcookbook::heightweight) +
  labs(color = "Male/Female", shape = "Male/Female")
```

## 10.6 Changing the Appearance of a Legend Title

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  theme(legend.title = element_text(color = "red", face = "italic", size = rel(1.5)))

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  guides(fill = guide_legend(title.theme = element_text(color = "red", face = "italic", size = 15)))
```

## 10.7 Removing a Legend Title

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  theme(legend.title = element_blank())

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  guides(fill = guide_legend(title = NULL))
```

## 10.8 Changing the Labels in a Legend

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  scale_fill_discrete(labels = c("ctrl" = "Control", "trt1" = "Treatment 1", "trt2" = "Treatment 2"))

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  scale_fill_discrete(
    labels = c("Treatment 1", "Treatment 2", "Control"),
    limits = c("trt1", "trt2", "ctrl")
  )
```

```r
qplot(ageYear, heightIn, color = sex, shape = sex, data = gcookbook::heightweight) +
  scale_color_discrete(labels = c("Female", "Male")) +
  scale_shape_discrete(labels = c("Female", "Male")) +
  labs(color = "Male/Female", shape = "Male/Female")
```

## 10.9 Changing the Appearance of Legend Labels

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  theme(legend.text = element_text(color = "red", face = "italic"))

qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  guides(fill = guide_legend(label.theme = element_text(color = "red", face = "italic")))
```

## 10.10 Using Labels with Multiple Lines of Text

```r
qplot(group, weight, fill = group, data = PlantGrowth, geom = "boxplot") +
  scale_fill_discrete(
    labels = c(
      "ctrl" = "Control",
      "trt1" = "Type 1\ntreatment",
      "trt2" = "Type 2\ntreatment"
    )
  ) +
  theme(
    legend.key.height = unit(1, "cm"),
    legend.text = element_text(lineheight = .75)
  )
```
